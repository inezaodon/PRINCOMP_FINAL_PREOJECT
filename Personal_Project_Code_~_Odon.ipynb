```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "yellow_tripdata_2015-01.csv"

df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "elemento/nyc-yellow-taxi-trip-data",
  file_path,)

print("First 5 records:", df.head())
```

## One Sample Visualization

```python
import plotly.express as px
import pandas as pd

# Assume df is loaded, if not, uncomment the placeholder below
# df = pd.read_csv('your_taxi_data.csv')

# --- DATA PREPARATION (from your original code) ---
df['pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Calculate trip duration in seconds
df['trip_duration'] = (df['dropoff_datetime'] - df['pickup_datetime']).dt.total_seconds()

# Filter out extreme outliers (less than 5 hours, 18000 seconds is 5 hours, not 1 hour as noted in original title)
# For clearer visualization, 18000 seconds (5 hours) is used as per your filtering,
# but a duration of 3600 seconds (1 hour) is often a better filter for short trips.
df_filtered = df[df['trip_duration'] < 18000].copy()

# --- VIOLIN PLOT GENERATION ---

# Create a dummy variable for the x-axis to plot a single distribution
df_filtered['overall_distribution'] = 'All Trips'

fig0_violin = px.violin(
    df_filtered,
    y='trip_duration',
    x='overall_distribution', # Use the dummy variable for a single plot
    box=True, # Show a box plot inside the violin for median and quartiles
    points=False, # Do not show individual data points
    title='Distribution of Trip Duration (Violin Plot)',
    labels={'trip_duration': 'Trip Duration (Seconds)', 'overall_distribution': ''},
    template='plotly_white'
)

fig0_violin.show()
```

```python
import pandas as pd
import plotly.express as px

# 1. DATA PRE-REQUISITES (MUST be executed first)
# *** NOTE: This assumes your dataframe 'df' is already loaded and cleaned ***

# Calculate Trip Speed (MPH)
# Speed = Distance (miles) / Duration (hours)
df['trip_speed_mph'] = (df['trip_distance'] / df['trip_duration_min']) * 60

# Data Cleaning: Filter out unrealistic speeds (e.g., > 60 MPH or near 0 MPH)
df_clean_speed = df[(df['trip_speed_mph'] > 1) & (df['trip_speed_mph'] < 60)].copy()

# Define Day Order for correct plotting
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_clean_speed['pickup_day_of_week'] = pd.Categorical(df_clean_speed['pickup_day_of_week'], categories=day_order, ordered=True)


# 2. V10: GENERATE HISTOGRAM

# V10: Distribution of Trip Speed (MPH) by Day of Week (REVISED for Congestion Analysis)
fig10 = px.histogram(df_clean_speed, x='trip_speed_mph', color='pickup_day_of_week',
                             title='V10: Distribution of Trip Speed (MPH) by Day of Week (Congestion Analysis)',
                             labels={'trip_speed_mph': 'Trip Speed (Miles Per Hour)', 'count': 'Number of Trips'},
                             marginal='box',  # Adds a box plot summary for overall speed
                             nbins=50,
                             template='plotly_white')
fig10.show()
```
